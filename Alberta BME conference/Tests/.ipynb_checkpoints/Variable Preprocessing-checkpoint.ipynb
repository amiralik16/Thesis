```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
```

```python
#Load the graph variables and Demographic variables
Dosenbach_path = os.getcwd() + '\\Dosenbach'
AAL_path = os.getcwd() + '\\AAL'

#Load Dosenbach
filelist = os.listdir(Dosenbach_path)
Dosenbach_df = pd.read_excel(Dosenbach_path + '\\globalVariables.xls',sheetname='0.2')
Dosenbach_df.columns = 'globalVariables' + '|' + Dosenbach_df.columns
for i in range(len(filelist)):
    if filelist[i] != 'globalVariables.xls':
        temp2 = pd.read_excel(Dosenbach_path + '\\' + filelist[i],sheetname='0.2')
        temp2 = temp2.drop('Unnamed: 160',axis=1)
        temp2.columns = [filelist[i][:-4] + '|'] + temp2.columns
        Dosenbach_df = Dosenbach_df.join(temp2)
    else:
        continue
del temp2
        
#Load AAL
filelist = os.listdir(AAL_path)
AAL_df = pd.read_excel(AAL_path + '\\globalVariables.xls',sheetname='0.2')
AAL_df.columns = 'globalVariables' + '|' + AAL_df.columns
for i in range(len(filelist)):
    if filelist[i] != 'globalVariables.xls':
        temp2 = pd.read_excel(AAL_path + '\\' + filelist[i],sheetname='0.2')
        temp2.columns = [filelist[i][:-4] + '|'] + temp2.columns
        AAL_df = AAL_df.join(temp2)
    else:
        continue
del temp2

Vars = pd.read_csv('Variables_Prepd.csv')
```

```python
def name_parser(s):
    '''
    pass
    '''
    b = s.replace('CorrMatrix_','')
    b = b[:b.find('_rois')]
    return b

def add_meta_data(df):
    '''
    pass
    '''
    temp_df = df.reset_index()
    temp_df['FILE_ID'] = temp_df['index'].apply(name_parser)
    temp_df['SITE_ID'] = temp_df['FILE_ID'].apply(lambda s: ''.join(s.split('_')[:-1]).upper())
    temp_df['subject'] = temp_df['FILE_ID'].apply(lambda s: s[s.find('_00')+3:])
    temp_df.set_index('index',inplace=True)
    return temp_df
    
```

```python
#Parse the index to add the meta data
Dosenbach_df = add_meta_data(Dosenbach_df)
AAL_df = add_meta_data(AAL_df)
```

```python
#Add usefull demographic data to the dataframes
Dosenbach_df = pd.merge(Dosenbach_df,Vars[['FILE_ID','DX_GROUP','SEX', 'AGE_AT_SCAN']],on='FILE_ID')
AAL_df = pd.merge(AAL_df,Vars[['FILE_ID','DX_GROUP','SEX', 'AGE_AT_SCAN']],on='FILE_ID')
```

```python
def select_subjects(df, sites=None, sex=None, age_range=None, group=None):
    '''
    pass
    '''
    
    dictionary = d = {'m': 1, 'f': 2, 'ASD': 1, 'HC': 2} #dict for transforming the inputs to values found in the df
    #Fill the variables in such a way that it selects all subjects
    if sites == None: 
        sites = list(df.SITE_ID.unique())
    if sex == None:
        sex = ['m','f']
    if age_range == None:
        age_range = (0,150)
    if group == None:
        group = ['HC','ASD']
    
    #Actual code
    if isinstance(group, list) == False:
        group = group.split()
    sex = list(map(dictionary.get,sex))
    group = list(map(dictionary.get,group))
    
    return df[(df['SITE_ID'].isin(sites)) & 
              (df['SEX'].isin(sex)) & 
              (df['AGE_AT_SCAN'].between(age_range[0], age_range[1])) & 
              (df['DX_GROUP'].isin(group)) 
             ]
```

```python
data = select_subjects(Dosenbach_df,sex='m',sites=['YALE','PITT','NYU','TRINITY','STANFORD','UCLA1','SDSU','UM1'])
X = data.drop(['DX_GROUP','SEX','AGE_AT_SCAN','FILE_ID','SITE_ID','subject'],axis=1)
y = data['DX_GROUP']
```

```python
from sklearn.preprocessing import MinMaxScaler
from sklearn.feature_selection import RFE,RFECV
from sklearn.pipeline import make_pipeline
from sklearn.model_selection import KFold
cv = KFold(n_splits=10,random_state=101)
```

```python
from sklearn.model_selection import cross_val_score,cross_validate,permutation_test_score,validation_curve
from sklearn.svm import SVC,LinearSVC

svc_rbf = make_pipeline(MinMaxScaler() ,SVC(random_state = 101,C=10,gamma='auto'))
svc_lin = LinearSVC(random_state=101)

print(f"RBF kernel score is {sum(cross_val_score(svc_rbf,X,y,cv=cv))/10}")
print(f"Linear kernel score is {sum(cross_val_score(svc_lin,X,y,cv=cv))/10}")
```

** Select by recursive feature selection **

```python
select = RFECV(SVC(kernel='linear',random_state=101),step=0.05,verbose=3,cv = cv, n_jobs=-1, scoring='accuracy')
```

```python
select.fit(X=X,y=y)
```

```python
select.score(X,y)
```

```python
select.n_features_
```

```python
len(select.grid_scores_)
```

```python
fig,ax = plt.subplots()
fig.set_size_inches(15,8)
plt.plot(range(1, len(select.grid_scores_) + 1), select.grid_scores_)
plt.show()
```

```python
(cross_val_score(SVC(kernel='linear',random_state=101),X.iloc[:,select.support_],y,cv=cv)).mean()
```

** Select features by a random forest **

```python
from sklearn.model_selection import cross_val_score,cross_validate,permutation_test_score,validation_curve
from sklearn.svm import SVC,LinearSVC
from sklearn.feature_selection import SelectFromModel
from sklearn.ensemble import RandomForestClassifier

lsvc = LinearSVC(C=10, penalty="l1", dual=False)
svc_rbf = make_pipeline(MinMaxScaler() ,SelectFromModel(lsvc), 
                        SVC(random_state = 101,C=10,gamma='auto'))
svc_lin = LinearSVC(random_state=101)

print(f"RBF kernel score is {sum(cross_val_score(svc_rbf,X,y,cv=cv))/10}")
print(f"Linear kernel score is {sum(cross_val_score(svc_lin,X,y,cv=cv))/10}")
```

```python
rf = RandomForestClassifier(n_estimators=1000,oob_score=True,n_jobs=-1,random_state=101,criterion='entropy')
```

```python
rf.fit(X,y)
```

```python
X.iloc[:,rf.feature_importances_>0.002]
```

```python
(cross_val_score(SVC(kernel='linear',random_state=101),X.iloc[:,rf.feature_importances_>0.001],y,cv=cv)).mean()
```

```python
(cross_val_score(SVC(kernel='rbf',random_state=101),X.iloc[:,rf.feature_importances_>0.002],y,cv=cv)).mean()
```

** Feature selection with Mlxtend SFS **

```python
from mlxtend.feature_selection import SequentialFeatureSelector as SFS

svc_rbf = make_pipeline(MinMaxScaler() ,SVC(random_state = 101,C=10,gamma='auto'))

sfs1 = SFS(svc_rbf, 
           k_features=(1,20), 
           forward=True, 
           floating=False, 
           verbose=2,
           scoring='accuracy',
           cv=cv,n_jobs=-1)

sfs1 = sfs1.fit(X, y)
```

```python
from mlxtend.feature_selection import SequentialFeatureSelector as SFS

svc_rbf = make_pipeline(MinMaxScaler() ,SVC(random_state = 101,C=10,gamma='auto'))

sfs2 = SFS(svc_rbf, 
           k_features=(20), 
           forward=True, 
           floating=True, 
           verbose=2,
           scoring='accuracy',
           cv=cv,n_jobs=-1)

sfs2 = sfs2.fit(X.values, y.values)
```

```python
from sklearn.preprocessing import StandardScaler

svc_rbf2 = make_pipeline(StandardScaler() ,SVC(random_state = 101,C=10,gamma='auto'))

sfs3 = SFS(svc_rbf2, 
           k_features=(1,40), 
           forward=True, 
           floating=False, 
           verbose=2,
           scoring='accuracy',
           cv=cv,n_jobs=-1)

sfs3 = sfs3.fit(X, y)
```
