```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import pyplot
import seaborn as sns
import pandas as pd
```

# Visualizations
In this notebook the code to visualize the age and sex distribution of subjects based on site

## Age

```python
filename = 'Variables_Prepd.csv'
df = pd.read_csv(filename)
df.head()
```

```python
df.info()
```

```python
sns.set_style('whitegrid')
sns.set_palette('deep')
```

**Plotting overal age distribution**

```python
sns.distplot(df['AGE_AT_SCAN'])
plt.show()
sns.boxplot(df['AGE_AT_SCAN'])
plt.show()
```

```python
g = sns.FacetGrid(data = df, hue = 'DX_GROUP',size=5,ylim=(0,0.1))
g.map(sns.distplot,'AGE_AT_SCAN', hist = False)
plt.show()
sns.boxplot(y = df['AGE_AT_SCAN'], x = df['DX_GROUP'])
plt.show()
```

**Plotting distribution based on imaging site**

```python
g = sns.FacetGrid(data = df,col ='SITE_ID', hue = 'DX_GROUP', col_wrap=4,ylim=(0,0.2),legend_out=True)
g.map(sns.distplot,'AGE_AT_SCAN', hist = False, rug = True).add_legend()
plt.show()
```

```python
fig, ax = pyplot.subplots()
fig.set_size_inches(10,6)
ax = sns.boxplot(x = 'SITE_ID', y = 'AGE_AT_SCAN', hue='DX_GROUP', data = df)
ax.set_xticklabels(ax.get_xticklabels(),rotation = 45)
plt.show()
```

## Sex
**Plotting male and female percentage and count for all subjects**

```python
sns.set(font_scale=1.2)
x=  'SEX'
y = 'Percentage'
hue = 'SITE_ID'
prop_df = (df[x]
           .groupby(df[hue])
           .value_counts(normalize=True)
           .rename(y)
           .reset_index())

fig, ax = plt.subplots()
fig.set_size_inches(15, 8.27)
g = sns.barplot(x=hue, y=y, hue=x, data=prop_df)
new_legend = ['Male','Female']
for t,l in zip(g.get_legend().texts,new_legend):
    t.set_text(l)
g.set_xticklabels(ax.get_xticklabels(),rotation = 45)
plt.show()
```

```python
sns.set(font_scale=1.2)
x=  'SEX'
y = 'Percentage'
hue = 'DX_GROUP'
prop_df = (df[x]
           .groupby(df[hue])
           .value_counts(normalize=True)
           .rename(y)
           .reset_index())

fig, ax = plt.subplots()
fig.set_size_inches(15, 8.27)
g = sns.barplot(x=hue, y=y, hue=x, data=prop_df)
new_legend = ['Male','Female']
for t,l in zip(g.get_legend().texts,new_legend):
    t.set_text(l)
g.set_xticklabels(['ASD','HC'])
plt.show()
```

```python
sns.set(font_scale=1.2)
x=  'SEX'
y = 'Count'
hue = 'SITE_ID'
prop_df = (df[x]
           .groupby(df[hue])
           .value_counts(normalize=False)
           .rename(y)
           .reset_index())

fig, ax = plt.subplots()
fig.set_size_inches(15, 8.27)
g = sns.barplot(x=hue, y=y, hue=x, data=prop_df)
new_legend = ['Male','Female']
for t,l in zip(g.get_legend().texts,new_legend):
    t.set_text(l)
g.set_xticklabels(ax.get_xticklabels(),rotation = 45)
plt.show()
```

**Plot num of each sex for each group**

```python
sns.countplot(x = 'DX_GROUP',hue = 'SEX', data = df)
plt.show()
```

```python
g = sns.FacetGrid(data = df,col ='SITE_ID', hue = 'SEX', col_wrap=4,legend_out=True, palette='Set2')
ax = g.map(sns.countplot, 'DX_GROUP', label = ['Male','Female']).add_legend()
plt.show()
```
